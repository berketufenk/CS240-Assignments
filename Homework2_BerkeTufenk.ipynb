```python
import numpy as np
import nsfg
import first
import thinkstats2
import thinkplot
%matplotlib inline
```

```python
resp=nsfg.ReadFemResp()
```

```python
#Q1
school=resp[resp.hieduc<=10]
university=resp[resp.hieduc>10]
```

```python
print ("Stats1= Mother age less than 18 years.")
print ("Stats2= Mother age between 18-19 years.")
print ("Stats3= Mother age between 20-24 years.")
print ("Stats4= Mother age between 25-29 years.")
print ("Stats5= Mother age between 30 or older years.")
print ("Stats96= Mother have no children")
```

```python
#Q2
university_pmf=thinkstats2.Pmf(university.agemomb1)
school_pmf=thinkstats2.Pmf(school.agemomb1)
```

```python
thinkplot.PrePlot(2)
thinkplot.Pmf(university_pmf,width=0.5, label="University", color="#00FF00")
thinkplot.Pmf(school_pmf, width=0.5, label="School", color="#FF0000")
thinkplot.Config(title="Agemomb1", xlabel="mother ages", ylabel="PMF")
```

```python
university_cdf=thinkstats2.Cdf(university.agemomb1)
school_cdf=thinkstats2.Cdf(school.agemomb1)
```

```python
thinkplot.PrePlot(2)
thinkplot.Cdfs([university_cdf, school_cdf])
thinkplot.Show(title="Agemomb1", xlabel="mother ages", ylabel="CDF")
```

```python
#Q3
list=[]
for i in school_pmf:
    res=100*school_pmf.Prob(i)-100*university_pmf.Prob(i)
    list.append(res)
thinkplot.Bar(school_pmf,list)
thinkplot.Config(title="Result of Differences btw group ages", xlabel="mothers age", ylabel="differences")
```

```python
#Q4
thinkplot.PrePlot(2)
thinkplot.Cdfs([university_cdf, school_cdf])
thinkplot.Show(title="Agemomb1", xlabel="mother ages", ylabel="CDF")
```

```python
underage18_uni=university_pmf.Prob(1)
underage18_sch=school_pmf.Prob(1)
perc_uni=1-underage18_uni
perc_sch=1-underage18_sch
```

```python
print "Probability of mother less than 18 in university:" +str(int(underage18_uni*100))
print "Probability of mother less than 18 in school:"+ str(int(underage18_sch*100))
print "Probability of mother greater than 18 in university:" +str(int(perc_uni*100))
print "Probability of mother greater than 18 in school:" + str(int(perc_sch*100))
```

```python
fmarno_db=resp.fmarno.value_counts()
fmarno_db
```

```python
marriage_number=thinkstats2.Hist(resp.fmarno, label="Number of Marriage")
marriage_number
```

```python
thinkplot.Hist(marriage_number)
thinkplot.Config(xlabel="Number of Marriages", ylabel="Respondents")
```

```python
neverMarried=resp[resp.fmarno==0]
married=resp[resp.fmarno!=0]
neverMarried
```

```python
married
```

```python
never_married_hist=thinkstats2.Hist(neverMarried.totincr, label=" Never Married Income")
married_hist=thinkstats2.Hist(married.totincr, label="Married Income")
```

```python
width=0.66
thinkplot.Hist(never_married_hist, align="right",width=width)
thinkplot.Hist(married_hist, align="left",width=width)
thinkplot.Config(xlabel="Incomes",ylabel="Families")
```

```python
print "Minimum value of Never Married: " + str(neverMarried.totincr.min()) + " Minimum value of Married: " + str(married.totincr.min()) + " Maximum value of Never Married: "+ str(neverMarried.totincr.max())+ " Maximum value of Married: "+ str(married.totincr.max())    
```

```python
print "Mean of Never Married: " + str(neverMarried.totincr.mean())+ " Mean of Married: " + str(married.totincr.mean())
```

```python
print " Standard Deviation of Never Married: " + str(neverMarried.totincr.std())+ " Standard Deviation of Married: " + str(married.totincr.std())
```

```python
print " Variance of Never Married: "+ str(neverMarried.totincr.var())+ " Variance of Married: "+ str(married.totincr.var())
```
